```python
import pandas as pd
import numpy as np

# 시드값 설정
np.random.seed(0)

# 데이터 프레임 크기 설정
num_rows = 100000

# 이름에 대한 확률 지정
name_probabilities = {'김철수': 0.3, '나민영': 0.4, '도민수': 0.3}

# 가구에 대한 확률 지정
fur_probabilities = {'책상': 0.5, '의자': 0.3, '받침대': 0.2}

# 휴가일수
vac_probabilities = {'15일': 0.8, '15일 이상 20일 미만': 0.15, '20일 이상': 0.05}

# 데이터 프레임 생성
names = np.random.choice(list(name_probabilities.keys()), size=num_rows, p=list(name_probabilities.values()))
furniture = np.random.choice(list(fur_probabilities.keys()), size=num_rows, p=list(fur_probabilities.values()))
vac_probabilities = np.random.choice(list(vac_probabilities.keys()), size=num_rows, p=list(vac_probabilities.values()))

df = pd.DataFrame({'이름': names, '사무용 가구': furniture, '휴가일수' : vac_probabilities})

# 데이터 프레임 출력
print(df)
```
