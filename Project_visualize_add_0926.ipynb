```python
from matplotlib import rc, font_manager as fm

# 나눔고딕 폰트 경로 설정
font_location = 'data/나눔고딕/NanumFontSetup_TTF_GOTHIC/NanumGothic.ttf'
font_name = fm.FontProperties(fname=font_location).get_name()

# 폰트 설정
rc('font', family=font_name)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 엑셀 파일 읽기
df = pd.read_excel('data/data3.xlsx')  # 파일 경로와 이름을 실제 파일의 경로와 이름으로 변경하세요.
# 데이터 확인
df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

df.plot(kind='bar', x='지역구', y=['개업수', '폐업수'], ax=ax)
plt.title('지역구별 개업수와 폐업수')
plt.xlabel('지역구')
plt.ylabel('수')
plt.xticks(rotation=45)
plt.show()
```

```python
# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 6))

# 막대 그래프 그리기
df.plot(kind='bar', x='지역구', y=['개업수', '폐업수'], ax=ax1, color=['lightblue', 'orange'])
ax1.set_xlabel('지역구')
ax1.set_ylabel('수')
ax1.set_xticklabels(df['지역구'], rotation=45)

# 꺾은선 그래프 추가하기
ax2 = ax1.twinx()
df.plot(kind='line', x='지역구', y='비율', ax=ax2, color='red')
ax2.set_ylabel('비율')
ax2.set_ylim(0, max(df['비율']) + 0.5)

plt.title('지역구별 개업수, 폐업수 및 비율')
plt.show()
```
