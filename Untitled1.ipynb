```python
import geopandas as gpd
import pandas as pd

# Shapefile 경로 지정
shp_file_path = 'data/2023상권영역/TBGIS_TRDAR_RELM.shp'

# Shapefile 불러오기
gdf = gpd.read_file(shp_file_path)

# 좌표값을 저장할 빈 리스트 생성
coordinates_list = []

# 각 폴리곤의 좌표값을 리스트에 추가
for index, row in gdf.iterrows():
    geometry = row['geometry']
    if geometry.geom_type == 'Polygon':
        coordinates = list(geometry.exterior.coords)
        coordinates_list.append(coordinates)
```

```python
# 아무리봐도 저렇게 불러온 걸 정리해서 불러왔다고 하긴 좀 그렇잖아?

# Shapefile 경로 지정
shp_file_path = 'data/2023상권영역/TBGIS_TRDAR_RELM.shp'

# Shapefile 불러오기
gdf = gpd.read_file(shp_file_path)

# CSV 파일로 저장
csv_file_path = 'data/2023상권영역/TBGIS_TRDAR_RELM.csv'
gdf.to_csv(csv_file_path, index=False)
```

```python
whkvy1 = whkvy[whkvy['SIGNGU_CD'] == 11680]
whkvy1
```

```python
whkvy1 = whkvy1[['TRDAR_CD', 'TRDAR_CD_N', 'geometry']]
whkvy1
```

```python
# merge 전 칼럼명 바꾸기
whkvy1 = whkvy1.rename(columns={
    'TRDAR_CD': '상권_코드',
    'TRDAR_CD_N': '상권_이름',
    'geometry': '폴리곤_좌표'
})
whkvy1
```

```python
whkvy2 = whkvy1[['상권_이름', '폴리곤_좌표']]
whkvy2
```

```python
temp = pd.read_csv('data/상권코드,상권명,시군구.csv')
temp
```

```python
temp = temp[['상권_코드', '상권_코드_명']]
temp
```

```python
merged_df = whkvy2.merge(temp, how='left', left_on='상권_이름', right_on='상권_코드_명')
merged_df
```

```python
whkvy3 = merged_df[['상권_코드', '폴리곤_좌표']]
whkvy3
```

```python
# df1과 whkvy3 데이터프레임을 레프트 조인
df_all = df1.merge(whkvy3, how='left', left_on='상권_코드', right_on='폴리곤_좌표')
df_all
```
