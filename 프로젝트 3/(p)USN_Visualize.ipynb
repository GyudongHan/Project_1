# 액체 비액체의 대기율 차이?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 조건에 따라 데이터 분리
condition = df['선박용도'].isin(['석유제품 운반선', '케미칼 운반선', '기타 유조선', 'LPG 운반선', '원유운반선', '케미칼가스 운반선', 'LNG 운반선'])
selected_rows = df[condition]
non_selected_rows = df[~condition]

# 각각의 평균 계산
selected_mean = selected_rows['내_대기율'].mean()
non_selected_mean = non_selected_rows['내_대기율'].mean()

# 시각화
plt.figure(figsize=(5, 3))
sns.barplot(x=['액체', 'Non_액체'], y=[selected_mean, non_selected_mean])
plt.title('대기율 비교')
plt.ylabel('대기율')
plt.show()
```

```python
# 선박용도별 대기율의 평균 계산
ship_WR = df.groupby('선박용도')['내_대기율'].mean().reset_index()
ship_WR.sort_values(by='내_대기율', ascending=False)
```

```python
# 시각화
plt.figure(figsize=(20, 10))
ax = sns.barplot(x='선박용도', y='내_대기율', data=ship_WR)
plt.title('평균대기율 by Ship Category')
plt.xticks(rotation=45, ha='right')  # X 축 라벨을 45도 기울임
plt.ylabel('평균대기율')

# 각 막대 위에 수치 표시
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

## 울산항 액체화물 / 비액체화물 대기율 비교

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 조건에 따라 데이터 분리
condition = df['선박용도'].isin(['석유제품 운반선', '케미칼 운반선', '기타 유조선', 'LPG 운반선', '원유운반선', '케미칼가스 운반선', 'LNG 운반선'])
selected_rows = df[condition]
non_selected_rows = df[~condition]

# 각각의 평균 계산
selected_mean = selected_rows['내_대기율'].mean()
non_selected_mean = non_selected_rows['내_대기율'].mean()

# 시각화
plt.figure(figsize=(7, 4))
sns.barplot(x=['액체', 'Non_액체'], y=[selected_mean, non_selected_mean])
plt.title('대기율 비교')
plt.ylabel('대기율')
plt.show()
```

## 액체화물 대기율을 줄이는 것이 울산항 전체적으로 효과적인 이유

```python
print(selected_rows.info())
print(selected_rows.describe())
```

```python
## 액체들의 접안 대기시간을 10% 줄여보자
```

```python
selected_rows['접안_대기시간_분'] = pd.to_numeric(selected_rows['접안_대기시간_분'], errors='coerce')

selected_rows['접안_대기시간_분'] = selected_rows['접안_대기시간_분'] * 0.6
selected_rows.describe()
```

```python
selected_rows['수정_대기율'] = selected_rows['접안_대기시간_분'] / selected_rows['Service_Time_분']
```

```python
selected_rows.describe()
```

```python
## NON액체도 그대로긴 한데 수정대기율 칼럼 만들어주기
```

```python
non_selected_rows['수정_대기율'] = non_selected_rows['접안_대기시간_분'] / non_selected_rows['Service_Time_분']
```

```python
non_selected_rows.describe()
```

## 컨캣해서 전체 울산항만 대기율 다시구해보기 차이가 어떨지?

```python
result_df = pd.concat([selected_rows, non_selected_rows], ignore_index=True)
result_df.describe()
```

```python
# '대기율' 컬럼의 평균값 계산
mean_df = df['내_대기율'].mean()*100

mean_df2 = result_df['수정_대기율'].mean()*100

# 정확한 수치 표시
plt.text(0, mean_df, f'{mean_df:.3f}', ha='center', va='bottom', color='black')
plt.text(1, mean_df2, f'{mean_df2:.3f}', ha='center', va='bottom', color='black')

# 막대그래프로 시각화
plt.bar(['대기시간개선_전', '대기시간개선_후'], [mean_df, mean_df2], color=['blue', 'orange'])
plt.title('대기율 평균값 비교')
plt.xlabel('데이터프레임')
plt.ylabel('평균값')
plt.show()
```
