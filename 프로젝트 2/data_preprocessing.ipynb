```python
data2 = data1[data1['업태_분류'] != '식품등 수입판매업']
data2.info()
```

```python
data3 = data2[data2['업태_분류'] != '건강기능식품수입업']
data3
```

```python
null_rows = data3[data3['행정동'].isnull()]
null_rows
```

```python
result2 = pd.merge(data3, df4, left_on='행정동', right_on='ADSTRD_NM', how='left')
result2
```

```python
result2[result2['ADSTRD_NM'].isnull()]
```

```python
# 데이터프레임의 전체 길이를 확인합니다.
total_length = len(result2)

# 나눌 크기를 정합니다.
chunk_size = 955000

# 데이터를 나누어 저장합니다.
for i in range(0, total_length, chunk_size):
    result_chunk = result2.iloc[i:i+chunk_size]
    
    # 파일 이름을 지정합니다. 예: chunk_1.csv, chunk_2.csv, ...
    file_name = f'chunk_{i//chunk_size + 1}.csv'
    
    # CSV 파일로 저장합니다.
    result_chunk.to_csv(file_name, encoding='utf-8-sig', index=False)
```
