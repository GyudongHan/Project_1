# 이것을 우리는 산업혁명 이후의 가장 큰 혁명으로 부르기로 했어요.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
df = pd.read_csv('data/구별_클러스터링용.csv')

subset = df.iloc[:, 1:]

# 'PET 총량' 변수와의 상관 관계가 0.4 이상인 변수
correlation_threshold = 0.4
correlation_with_pet = subset.corr()['PET 총량']
selected_variables = [col for col in correlation_with_pet.index if abs(correlation_with_pet[col]) >= correlation_threshold]

# 선택된 변수들 간의 상관 관계 히트맵
df_selected = df[selected_variables]
correlation_matrix = df_selected.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('선택된 변수들 간의 상관 관계 히트맵 (상관 관계 >= 0.4)')
plt.show()
```

# 랜덤포레스트 결정트리 코드

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import plotly.express as px


# 데이터 불러오기
dfr = pd.read_csv('data/for_k_means.csv')
dfr.drop(columns='자치구', inplace=True)
dfr
```

```python
dfr.describe()
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# PET 총량 칼럼 데이터
pet_total = dfr['PET 총량']

# Figure 및 서브플롯 설정
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# 1번 그래프: 히스토그램
ax1.hist(pet_total, bins=30, color='blue', alpha=0.7, edgecolor='black')
ax1.set_title('히스토그램')
ax1.set_xlabel('PET 총량')
ax1.set_ylabel('빈도수')

# 왜도 (skewness) 계산
skew_value = stats.skew(pet_total)
ax1.text(0.6, 0.9, f'Skewness: {skew_value:.2f}', transform=ax1.transAxes, fontsize=12)

# 2번 그래프: Box plot
ax2.boxplot(pet_total, vert=False)
ax2.set_title('Box plot')
ax2.set_yticklabels(['PET 총량'])  # y축 레이블 설정

# 레이아웃 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

### 데이터의 크기: 
- 대규모 데이터셋: 대부분의 경우, 대규모 데이터셋에서는 두 검정 중 어떤 것을 선택하더라도 정확한 결과를 얻을 수 있습니다.
- 소규모 데이터셋: 작은 크기의 데이터셋인 경우, 샤피로-윌크 검정이 더 신뢰할 만합니다. 이는 콜모고로프-스미르노프 검정이 큰 데이터셋에서 미세한 차이를 감지하기 때문입니다. 
### 분포의 형태: 
- 이론적으로 정규분포에 가까운 데이터: 콜모고로프-스미르노프 검정을 사용할 수 있습니다.
- 비대칭적이거나 이상치가 있는 데이터: 샤피로-윌크 검정이 더 강건한(robust) 특성을 가지므로, 이러한 상황에서 신뢰할  수 있습니다.

```python
from scipy.stats import shapiro
data = dfr["PET 총량"]

# data는 검정을 수행할 데이터입니다.
statistic, p_value = shapiro(data)

# p-value를 통해 검정 결과를 해석합니다.
if p_value > 0.05:
    print("데이터는 정규분포를 따릅니다.")
    print(p_value)
else:
    print("데이터는 정규분포를 따르지 않습니다.")
    print(p_value)
```

```python
# 사용할 컬럼 선택
y = dfr.iloc[:, 0]
X = dfr.iloc[:, 6:]

# 데이터 분할 (random_state 설정)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)  # random_state 추가

# n_estimators 범위
n_estimators_range = [10, 50, 100, 200, 300, 400, 450, 475, 500]
best_n_estimators = None
best_mae = float('inf')  # MAE 최소화

# n_estimators 범위 평가 및 MAE 기록
mae_scores = []
for n_estimators in n_estimators_range:
    model = RandomForestRegressor(n_estimators=n_estimators, max_depth=5, random_state=0)  # random_state 추가
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mae_scores.append(mae)
    
    # 현재의 n_estimators 값이 이전보다 더 좋으면 최적값 업데이트
    if mae < best_mae:
        best_mae = mae
        best_n_estimators = n_estimators

fig.show()

print(f"최적 n_estimators: {best_n_estimators}")
print(f"최적 MAE: {best_mae}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
# 사용할 컬럼 선택
y = dfr.iloc[:, 3]
X = dfr.iloc[:, 6:]

# 데이터 분할 (random_state 설정)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)  # random_state 추가

# n_estimators 범위
n_estimators_range = list(range(10, 501, 1))
best_n_estimators = None
best_mae = float('inf')  # MAE 최소화
best_rmse = None  # RMSE 기록 초기화

# n_estimators 범위 평가 및 MAE, RMSE 기록
mae_scores = []
rmse_scores = []  # RMSE 기록 리스트 추가
for n_estimators in n_estimators_range:
    model = RandomForestRegressor(n_estimators=n_estimators, max_depth=5, random_state=0)  # random_state 추가
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mae_scores.append(mae)
    rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE 계산
    rmse_scores.append(rmse)  # RMSE 기록
    
    # 현재의 n_estimators 값이 이전보다 더 좋으면 최적값 업데이트
    if mae < best_mae:
        best_mae = mae
        best_rmse = rmse  # 최적 RMSE 기록 추가
        best_n_estimators = n_estimators

print(f"최적 n_estimators: {best_n_estimators}")
print(f"최적 MAE: {best_mae}")
print(f"최적 RMSE: {best_rmse}")
```

```python
import numpy as np
# 사용할 컬럼 선택
y = dfr.iloc[:, 3]
X = dfr.iloc[:, 6:]

# 데이터 분할 (random_state 설정)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)  # random_state 추가

# 로그 변환
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)

# n_estimators 범위
n_estimators_range = [10, 50, 100, 200, 300, 400, 450, 475, 500]
best_n_estimators = None
best_mae = float('inf')  # MAE 최소화
best_rmse = None  # RMSE 기록 초기화

# n_estimators 범위 평가 및 MAE, RMSE 기록
mae_scores = []
rmse_scores = []  # RMSE 기록 리스트 추가
for n_estimators in n_estimators_range:
    model = RandomForestRegressor(n_estimators=n_estimators, max_depth=5, random_state=0)  # random_state 추가
    model.fit(X_train, y_train_log)
    y_pred_log = model.predict(X_test)
    y_pred = np.expm1(y_pred_log)  # 로그 변환된 예측값을 원래 스케일로 변환
    mae = mean_absolute_error(y_test, y_pred)
    mae_scores.append(mae)
    rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE 계산
    rmse_scores.append(rmse)  # RMSE 기록
    
    # 현재의 n_estimators 값이 이전보다 더 좋으면 최적값 업데이트
    if mae < best_mae:
        best_mae = mae
        best_rmse = rmse  # 최적 RMSE 기록 추가
        best_n_estimators = n_estimators

# 최적 결과 출력
print(f"최적 n_estimators: {best_n_estimators}")
print(f"최적 MAE: {best_mae}")
print(f"최적 RMSE: {best_rmse}")
```

```python
#만약 로그 변환을 했을 때 MAE가 증가한다면, 해당 데이터에 대해서는 로그 변환을 적용하지 않는 것이 더 나을 수 있습니다. 로그 변환은 모든 경우에 항상 성능을 향상시키지는 않습니다.

# 로그 변환은 주로 데이터의 분포가 지수적으로 증가하는 경우에 유용하며, 이렇게 하면 모델이 데이터의 비선형성을 더 잘 학습할 수 있습니다. 그러나 모든 데이터가 이런 특성을 가지는 것은 아닙니다.

# 따라서 데이터의 특성을 고려하고 실험을 통해 어떤 전처리가 더 나은 결과를 가져다주는지 확인하는 것이 중요합니다.MAE가 높아질 경우에는 로그 변환을 하지 않는 것이 오히려 더 좋은 선택일 수 있습니다.
```
