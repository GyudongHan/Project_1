```python
temp2_mod = temp2_mod.drop(columns=temp2_mod.columns[10:58], axis=1)
temp2_mod
```

```python
temp_group = temp2_mod.groupby(['연도', '월', '버스정류장ARS번호']).sum().reset_index()
temp_group
```

```python
# '월' 컬럼을 정수형으로 변환
temp_group['월'] = temp_group['월'].astype(int)

# 분기를 만드는 함수
def make_quarter(month):
    if month in [1, 2, 3]:
        return '1분기'
    elif month in [4, 5, 6]:
        return '2분기'
    elif month in [7, 8, 9]:
        return '3분기'
    elif month in [10, 11, 12]:
        return '4분기'

# 데이터프레임에 분기 컬럼 추가
temp_group['분기'] = temp_group['월'].apply(make_quarter)
temp_group
```

```python
# 월 컬럼 삭제
# 역명 컬럼 삭제
# 연도, X,Y 좌표 컬럼 삭제 : 빠진 이유는 이후 merge 할 때 더해지기 때문
# 컬럼순서 정의 
column_order = ['분기', '버스정류장ARS번호', '00~06_승하차_승객수',
                '06~11_승하차_승객수', '11~14_승하차_승객수', '14~17_승하차_승객수',
                '17~21_승하차_승객수', '21~24_승하차_승객수']

# 데이터프레임의 컬럼 순서를 재정렬
temp_group = temp_group[column_order]

temp_group
```

```python
# 분기별로 정류장 하나의 값들을 합해줌.
merged_data1 = temp_group.groupby(['버스정류장ARS번호', '분기'])[['00~06_승하차_승객수', '06~11_승하차_승객수', '11~14_승하차_승객수', '14~17_승하차_승객수', '17~21_승하차_승객수', '21~24_승하차_승객수']].sum().reset_index()
merged_data1
```

```python
# 아까 만들었던 temp2_mod에서 ARS 번호와 , X,Y 좌표만을 남긴 CSV 파일을 불러옴.
whkvy = pd.read_csv('data/bus_data/temp2_mod(XY).csv', encoding='cp949')

whkvy = whkvy.drop_duplicates()

whkvy
```

```python
result1 = pd.merge(merged_data1, whkvy, on="버스정류장ARS번호", how="right")

result1
```
